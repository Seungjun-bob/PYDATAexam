```python
import pandas as pd
```

```python
data = {
    'name':['둘리', '또치', '도우너', '희동이'],
    'kor':[90, 80, 70, 70],
    'eng':[99, 98, 97, 46],
    'mat':[90, 70, 70, 60],
}
df = pd.DataFrame(data)
```

```python
df['class'] = ['1반', '2반', '1반', '2반']
df
```

```python
df.loc[len(df)] = ['마이콜', 80, 80, 80 ,'1반']
df.set_index('name', inplace=True)
df
```

```python
df.rename(columns={'kor':'국어', 'eng':'영어', 'mat':'수학', 'class':'반번호'}, inplace=True)
df
```

```python
df.loc['마이콜', ['국어', '영어', '수학']] = 100
df.loc['희동이', '영어'] = 90
df
```

```python
df = df.reset_index()
df.rename(columns={'name':'성명'}, inplace=True)
df
```

```python
df1 = df.sort_values('국어', ascending=False)
df2 = df.sort_values('영어')
df1, df2
```

```python
import numpy as np
df['총점'] = df.sum(axis=1)
df
```

```python
df.sort_values('총점', ascending=False)
```

```python
import numpy as np
data1 = {'kor':90, 'eng':np.nan, 'mat':80}
data2 = {'kor':90, 'eng':70, 'mat':np.nan}
data3 = {'kor':90, 'eng':70, 'mat':80}
series1 = pd.Series( data1 )
series2 = pd.Series( data2 )
series3 = pd.Series( data3 )
result = series1.add(series2, fill_value=0)
result = result.add(series3, fill_value=0)
print(result)
```

```python
data = {
    'X1':[2.9, 2.4, 2, 2.3, 3.2],
    'X2':[9.2, 8.7, 7.2, 8.5, 9.6],
    'X3':[13.2, 11.5, 10.8, 12.3, 12.6],
    'X4':[2, 3, 4, 3, 2]
} 
df = pd.DataFrame(data, index=['Y1','Y2','Y3', 'Y4', 'Y5'])
print(df)
```

```python
new_data = {
    'X1':[10.0],
    'X2':[20.0],
    'X3':[30.0],
    'X4':[40]
} 
new_df = pd.DataFrame(new_data, index= ['Y6'])

df = pd.concat([df, new_df])
df
```
