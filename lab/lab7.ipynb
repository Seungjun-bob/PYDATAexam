```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
import time

url = 'http://gs25.gsretail.com/gscvs/ko/products/event-goods'
driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.get(url)

two_to_one_page = driver.find_element_by_css_selector("#TWO_TO_ONE")
two_to_one_page.click()
```

```python
goodsname = [] 
goodsprice = []

while True:
    name = driver.find_elements_by_css_selector("#contents > div.cnt > div.cnt_section.mt50 > div > div > div:nth-child(5) > ul > li > div > p.tit")
    price = driver.find_elements_by_css_selector("#contents > div.cnt > div.cnt_section.mt50 > div > div > div:nth-child(5) > ul > li > div > p.price > span")

    for i in range(len(name)) :
        goodsname.append(name[i].text)
        goodsprice.append(price[i].text)
        
    linkurl = '#contents > div.cnt > div.cnt_section.mt50 > div > div > div:nth-child(5) > div > a.next'
    try :
        linkNum = driver.find_element_by_css_selector(linkurl)
    except :
        print("더 이상 다음 페이지 없음")
        break
    linkNum.click()  
    time.sleep(1)
```

```python
temp  = []
import re
for dom in goodsprice:
    dom = re.sub("원",'',dom)
    dom = re.sub(",",'',dom)
    temp.append(dom)
goodsname
```

```python
with open('output/gs25_twotoone.csv', "wt", encoding="utf-8") as f:
    f.write('goodsname,goodsprice\n')  
    for i in range(len(goodsname)):
        f.write(goodsname[i]+","+temp[i]+'\n')  
```
